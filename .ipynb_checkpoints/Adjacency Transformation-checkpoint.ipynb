```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from lxml import objectify
import math
import urllib

# matplotlib
%matplotlib inline
pd.set_option('display.width', 5000)
pd.set_option('display.max_columns', 60) 
plt.style.use('default')
```

```python
# w_counts.to_csv(path_or_buf='NovemberWeekdayJourneys.csv')
november.df.Duration = november.df.Duration.astype(int)
sw_time = november.df.s_date.apply(lambda x: x.weekday() < 5)
w_counts = november.df.loc[sw_time, ['Duration', 'e_date', 'e_id', 's_date', 's_id']]
w_counts.set_index(w_counts.s_date, inplace=True)
w_counts.set_index(w_counts.index.time, inplace=True)
w_counts
```

```python
hours = pd.Series(pd.date_range(start='2015-01-01', periods=96, freq='15Min')).dt.time
peak_time = (w_counts.index > hours[32]) & (w_counts.index < hours[33])
am_peak = w_counts.loc[peak_time, ['s_id', 'e_id', 's_date', 'e_date', 'Duration']]
# am_peak.to_csv('AM_Peak.csv', header=['Source','Target','Start','End','Duration'], index=False)
am_peak.e_id = am_peak.e_id.astype(int)
am_peak.s_id = am_peak.s_id.astype(int)
am_peak.dtypes
# If Duration > x, use in subsequent time frame dataframe with key?
```

```python
counts = pd.Series()
adjacency = pd.DataFrame(index=pd.Series(range(1, 781)), columns=pd.Series(range(1, 781)))
for x in range(0,len(am_peak)):
    sid = am_peak.iloc[x][0]
    eid = am_peak.iloc[x][1]
    row_indexer = (am_peak.s_id == sid) & (am_peak.e_id == eid)
    multiples = am_peak.loc[row_indexer, ['s_id', 'e_id']]
    counts = counts.append(pd.Series([len(multiples)]))      
    adjacency[eid][sid] = len(multiples)
counts.reset_index(drop=True, inplace=True)
am_peak.reset_index(drop=True, inplace=True)
am_peak['counts'] = counts
```

```python
edges = adjacency.dropna([0,1], how='all').fillna(0)
edges = edges.astype(int)
edges.to_csv('EdgesMetrix(eidxsid).csv', sep=';')
```

```python
# All journeys from 287 - 232
rowspot = (am_peak.s_id == 287) & (am_peak.e_id == 232)
trip = am_peak.loc[rowspot, ['s_id', 'e_id', 'Duration', 'counts']]
trip.Duration.mean() # Potentially use in weightings
```
